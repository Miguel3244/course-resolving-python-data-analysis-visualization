## Create a scatter plot

### Import matplotlib library

```python
import matplotlib.pyplot as plt
```

```python
df_gapminder = pandas.read_excel('gapminder.xlsx')
```

### Access column data separately

```python
df_gapminder.money_per_person
```

```python
df_gapminder.life_expectancy
```

### Pass data to scatter function

```python
plt.scatter (x=df_gapminder.money_per_person ,y=df_gapminder.life_expectancy, c='red', edgecolors='black', s=80)
```

## Customize the plot

### Change the size

```python
plt.scatter ( x=df_gapminder.money_per_person, y = df_gapminder.life_expectancy, s = 75)
```

### Change the color

```python
plt.scatter (x=df_gapminder.money_per_person ,y=df_gapminder.life_expectancy, c='red', s=80)
```
